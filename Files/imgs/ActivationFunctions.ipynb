```python
import matplotlib.pyplot as plt
import numpy as np

# https://github.com/colah/nnftd/blob/master/fig/chap3/relu.py
# each of our activation functions needs a defined function
def ReLU(x):
    return np.max(0,x)

def sigmoid(x):
    return 1/(1+np.exp(-x))

# for jupyter notebook
%matplotlib inline

# rectified linear unit
x = np.arange(-2, 2, .1)
zero = np.zeros(len(x))
y = np.max([zero, x], axis=0)

fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(16,4),dpi=600)
ax1.plot(x,y,color='red')
ax1.set_ylim([-2.0, 2.0])
ax1.set_xlim([-2.0, 2.0])
ax1.axvline(x=0,color='black')
ax1.grid(True)
ax1.set_xlabel('x')
ax1.set_title(r'$max(0,x)$',fontsize=15,pad=10)
plt.savefig('ReLU.jpg',dpi=300)


# sigmoid
z = np.arange(-5, 5, .1)
sigma_fn = np.vectorize(lambda z: 1/(1+np.exp(-z)))
sigma = sigma_fn(z)

ax2.plot(z, sigma,color='blue')
ax2.set_ylim([0, 1])
ax2.set_xlim([-5,5])
ax2.axhline(y=0,color='black')
ax2.axvline(x=0,color='black')
ax2.grid(True)
ax2.set_xlabel('x')
ax2.set_title(r'$\frac{1}{1+e^{-x}}$',fontsize=20,pad=15)


# tanh
z = np.arange(-5, 5, .1)
t = np.tanh(z)
ax3.plot(z, t,color='blue')
ax3.set_ylim([-1.0, 1.0])
ax3.axhline(y=0,color='black')
ax3.axvline(x=0,color='black')
ax3.set_xlim([-4,4])
ax3.grid(True)
ax3.set_xlabel('x')
ax3.set_title(r'$tanh(x)$',fontsize=15,pad=10)

fig.savefig('activationfunctions.jpg',dpi=600)
```
