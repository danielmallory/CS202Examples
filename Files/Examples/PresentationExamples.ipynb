```python
# import statements
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# generic 'Sobel' filter used in edge detection
sobel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
# load in and greyscale image
img = mpimg.imread('cat.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# run Sobel filter over image
filtimg = cv2.filter2D(gray, -1, sobel)
```

```python
fig, ax = plt.subplots(1,3,figsize=(20,5),dpi=600)
ax[0].imshow(img)
ax[0].set_title('Pre-Filtered Image',fontsize=20)
ax[1].imshow(sobel)
ax[1].set_title('Sobel Filter', fontsize=20)
ax[1].set_xticks([0,1,2])
ax[1].set_yticks([0,1,2])
for i in range(3):
    for j in range(3):
        text = ax[1].text(j, i, sobel[i,j],
                       ha="center", va="center", color="red",fontsize=20)
ax[2].imshow(filtimg)
ax[2].set_title('Sobel-Filtered Image', fontsize=20)
fig.savefig('filterplots.jpg',dpi=600)
```
